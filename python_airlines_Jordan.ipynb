```python
# Welcome message and initial user inputs
print("Welcome to Python Airlines Prototype Membership System\n")

# Input for first name and flight data
firstName = input("Please enter your first name: ")
annualMiles = int(input("How many miles have you flown this year? "))
annualSegments = int(input("How many segments have you flown this year? "))
annualDollars = float(input("How much money have you spent this year (USD)? "))
lifetimeMiles = int(input("How many lifetime miles have you flown? "))

# Reward Variables Initialization
bags = 0
upgrades = ""
loungeAccess = None
discount = 0

# Input for upcoming flight's lounge check and ticket price
loungeCheck = input("Is your next flight to (or through) the Chicago airport? (yes/no): ").strip().lower()

if (loungeCheck in ["yes", "y"]):
    loungeCheck = True
else:
    loungeCheck = False

ticketPrice = float(input("How much will a Coach Class ticket for your next flight cost (USD)? "))

# Membership and Lifetime Membership Variables Initialization
membershipTierSapphire = False
membershipTierEmerald = False
membershipTierRuby = False
membershipTierDiamond = False
lifetimeMembershipEmerald = False
lifetimeMembershipRuby = False
lifetimeMembershipDiamond = False

# Lifetime Membership Determination
if (lifetimeMiles >= 3000000):
    lifetimeMembershipDiamond = True
    lifetimeMembershipRuby = True
    lifetimeMembershipEmerald = True
elif (lifetimeMiles >= 2000000):
    lifetimeMembershipRuby = True
    lifetimeMembershipEmerald = True
elif (lifetimeMiles >= 1000000):
    lifetimeMembershipEmerald = True

# Annual Membership Determination
if (annualMiles >= 100000 or (annualSegments >= 120 and annualDollars >= 24000)):
    membershipTierDiamond = True
    bags = 3
    upgrades = "First Class"
    loungeAccess = True
    discount = 0.30
elif (annualMiles >= 75000 or (annualSegments >= 90 and annualDollars >= 15000)):
    membershipTierRuby = True
    bags = 2
    upgrades = "First Class"
    loungeAccess = False
    discount = 0.25
elif (annualMiles >= 50000 or (annualSegments >= 60 and annualDollars >= 7000)):
    membershipTierEmerald = True
    bags = 1
    upgrades = "Business Class"
    loungeAccess = False
    discount = 0.20
elif (annualMiles >= 25000 or (annualSegments == 30 and annualDollars == 3000)):
    membershipTierSapphire = True
    upgrades = "Business Class"
    loungeAccess = True
    discount = 0.15

# Input Summary with Proper Formatting
print(f"\n{firstName}, this year, you have flown {annualSegments:,} flights for {annualMiles:,} miles "
      f"and spent ${annualDollars:,.2f} USD with Python Airlines. Lifetime, you have flown {lifetimeMiles:,} miles.\n")

# Lifetime Membership Status
if (lifetimeMembershipDiamond):
    print("You have achieved lifetime frequent-flier status at the Diamond Level!\n")
elif (lifetimeMembershipRuby):
    print("You have achieved lifetime frequent-flier status at the Ruby Level!\n")
elif (lifetimeMembershipEmerald):
    print("You have achieved lifetime frequent-flier status at the Emerald Level!\n")

# Annual Membership Status
if (membershipTierDiamond):
    print("This year, you have achieved frequent-flier status at the Diamond Level!\n")
elif (membershipTierRuby):
    print("This year, you have achieved frequent-flier status at the Ruby Level!\n")
elif (membershipTierEmerald):
    print("This year, you have achieved frequent-flier status at the Emerald Level!\n")
elif (membershipTierSapphire):
    print("This year, you have achieved frequent-flier status at the Sapphire Level!\n")
else:
    print("You are on your way to achieving annual frequent-flier status with Python Airlines!\n")

# Reward Summary
print("You have unlocked the following rewards:")

# Checked Bags Reward
print(f" - {bags} free checked bags per flight")

# Upgrade Reward
print(f" - Free seat upgrades to {upgrades}")

# Lounge Access Reward
if (loungeAccess):
    if (loungeCheck):
        print(" - Enjoy free access to the club lounge in Chicago on your next/upcoming flight!")
    else:
        print(" - Free access next time you fly to/through Chicago.")

# Discount Reward and Ticket Price Calculation
if (discount > 0):
    discountedPrice = ticketPrice * (1 - discount)
    print(f" - Your upcoming flight ticket price of ${ticketPrice:,.2f} was reduced by {int(discount*100)}% "
          f"to ${discountedPrice:,.2f}.")
```
